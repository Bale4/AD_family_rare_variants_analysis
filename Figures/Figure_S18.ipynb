# Figure S18: QQ plots of common variant associations (APOE4 adjustment)

Note: For the meta-analysis the p-value column is "P.value"

```sos
plots <- function(file,out){
library('qqman')
data <- read.table(gzfile(file), sep='\t', header=T)
lambda <- median(qchisq(1-data$P.value,1), na.rm=TRUE)/qchisq(0.5,1)
png(out, width = 5, height = 5, unit='in', res=300)
qq_plot <- qq(data$P.value, xlim = c(0, 8), ylim = c(0, 18), pch = 18, col = "blue4", cex = 1.5, las = 1)
text(3,13, paste("lambda","=",  signif(lambda, digits = 3)) )
dev.off()
}
```

```sos
# Create panel plots
library(patchwork)
library(png)
library(gridExtra)
library(grid)
library(ggplot2)
```

```sos
# File paths for your  PNG figures
file_paths <- c("~/project_bst/AD_family_Rare_variants/plots/GMMAT/AD_family_meta_qq_APOE4.png",
                "~/project_bst/AD_family_Rare_variants/plots/GMMAT/ADSP_meta_qq_APOE4.png",
               "~/project_bst/AD_family_Rare_variants/plots/GMMAT/AD_family_common_variants_NHW_APOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/GMMAT/ADSP_common_variants_NHW_APOE4.qq.png",
               "~/project_bst/AD_family_Rare_variants/plots/GMMAT/AD_family_common_variants_CH_APOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/GMMAT/ADSP_common_variants_CH_APOE4.qq.png",
               "~/project_bst/AD_family_Rare_variants/plots/GMMAT/AD_family_common_variants_AA_APOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/GMMAT/ADSP_common_variants_AA_APOE4.qq.png")

# Create a list to store individual plots
plots <- lapply(seq_along(file_paths), function(i) {
  # Read PNG file and create a ggplot
  ggplot() +
    annotation_custom(rasterGrob(readPNG(file_paths[i]), width = 1, height = 1), xmin=-Inf, xmax=Inf, ymin=-Inf, ymax=Inf) +
    theme_void() +  # Use theme_void to remove axis labels and ticks
    labs(title = letters[i])  # Add labels (A, B, C, etc.)
})

# Combine the individual plots into a single multiple-panel figure
combined_figure_all <- wrap_plots(plots, nrow = 4, ncol = 2)
ggsave("~/project_bst/AD_family_Rare_variants/plots/GMMAT/qq_APOE4.png", plot = combined_figure_all)
# Display the combined figure
combined_figure_all
```
