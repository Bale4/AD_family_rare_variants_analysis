# Figure S8 : QQ plots for pLoF rare-variant aggregate analysis of AD-family and ADSP data without APOE4 adjustment

```sos
plots <- function(file,out){
library('qqman')
data <- read.table(gzfile(file), sep='\t', header=T)
lambda <- median(qchisq(1-data$P,1), na.rm=TRUE)/qchisq(0.5,1)
png(out, width = 5, height = 5, unit='in', res=300)
qq_plot <- qq(data$P, xlim = c(0, 6), ylim = c(0, 12), pch = 18, col = "blue4", cex = 1.5, las = 1)
text(3,18, paste("lambda","=",  signif(lambda, digits = 3)) )
dev.off()
}
```

# Generate multi-panel figures

```sos
# Multi-panel figure for all AD-family and ADSP across ancestries meta-analysis
```

```sos
# Create panel plots
library(patchwork)
library(png)
library(gridExtra)
library(grid)
library(ggplot2)
```

```sos
# File paths for your  PNG figures
file_paths <- c("~/project_bst/AD_family_Rare_variants/plots/SKAT_O/AD_family_NHW_model1_noAPOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/ADSP_NHW_model1_noAPOE4.qq.png",
               "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/AD_family_CH_model1_noAPOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/ADSP_CH_model1_noAPOE4.qq.png",
               "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/AD_family_AA_model1_noAPOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/ADSP_AA_model1_noAPOE4.qq.png")

# Create a list to store individual plots
plots <- lapply(seq_along(file_paths), function(i) {
  # Read PNG file and create a ggplot
  ggplot() +
    annotation_custom(rasterGrob(readPNG(file_paths[i]), width = 1, height = 1), xmin=-Inf, xmax=Inf, ymin=-Inf, ymax=Inf) +
    theme_void() +  # Use theme_void to remove axis labels and ticks
    labs(title = letters[i])  # Add labels (A, B, C, etc.)
})

# Combine the individual plots into a single multiple-panel figure
combined_figure_all <- wrap_plots(plots, nrow = 3, ncol = 2)
ggsave("~/project_bst/AD_family_Rare_variants/plots/SKAT_O/per_population_model1_noAPOE4.qq.png", plot = combined_figure_all)
# Display the combined figure
combined_figure_all
```
