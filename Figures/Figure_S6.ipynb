# Figure S6 : QQ plots for rare-variant aggregate meta-analysis within ancestries of AD-family and ADSP data with APOE4 adjustment

```sos
plots <- function(file,out){
library('qqman')
data <- read.table(gzfile(file), sep='\t', header=T)
lambda <- median(qchisq(1-data$P.value,1), na.rm=TRUE)/qchisq(0.5,1)
png(out, width = 5, height = 5, unit='in', res=300)
qq_plot <- qq(data$P.value, xlim = c(0, 6), ylim = c(0, 30), pch = 18, col = "blue4", cex = 1.5, las = 1)
text(3,18, paste("lambda","=",  signif(lambda, digits = 3)) )
dev.off()
}
```

# Generate multi-panel figures

```sos
# Multi-panel figure for all AD-family and ADSP across ancestries meta-analysis
```

```sos
# Create panel plots
library(patchwork)
library(png)
library(gridExtra)
library(grid)
library(ggplot2)
```

```sos
# File paths for your  PNG figures
file_paths <- c("~/project_bst/AD_family_Rare_variants/plots/SKAT_O/European_meta_model1_APOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/European_meta_model2_APOE4.qq.png",
               "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/Hispanic_meta_model1_APOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/Hispanic_meta_model2_APOE4.qq.png",
               "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/African_meta_model1_APOE4.qq.png",
                "~/project_bst/AD_family_Rare_variants/plots/SKAT_O/African_meta_model2_APOE4.qq.png")

# Create a list to store individual plots
plots <- lapply(seq_along(file_paths), function(i) {
  # Read PNG file and create a ggplot
  ggplot() +
    annotation_custom(rasterGrob(readPNG(file_paths[i]), width = 1, height = 1), xmin=-Inf, xmax=Inf, ymin=-Inf, ymax=Inf) +
    theme_void() +  # Use theme_void to remove axis labels and ticks
    labs(title = letters[i])  # Add labels (A, B, C, etc.)
})

# Combine the individual plots into a single multiple-panel figure
combined_figure_all <- wrap_plots(plots, nrow = 3, ncol = 2)
ggsave("~/project_bst/AD_family_Rare_variants/plots/SKAT_O/within_ancestries_meta_APOE4.qq.png", plot = combined_figure_all)
# Display the combined figure
combined_figure_all
```
